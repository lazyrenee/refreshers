### Loading data

```python
import pandas as pd
import re
```

```python
df = pd.read_csv("pokemon_data.csv")
df_xlsx = pd.read_excel("pokemon_data.xlsx")
df_txt = pd.read_csv("pokemon_data.txt", sep = "\t")

print(df.head())
print(df_xlsx.head())
print(df_txt.head())
```

### reading data

```python
# reading headers
print(df.columns)

# read each column
print(df[["Name", "Type 1"]])

# read each row
print(df.iloc[0:4])


# read a specific location (r,c)
print(df.iloc[2,1])
```

### Sorting / describing data

```python
df.describe()
```

```python
df.sort_values("Name", ascending = False)
```

```python
df.sort_values(["Type 1", "HP"], ascending = True)
```

### making changes to data

```python
df["total"] = df["HP"] + df["Attack"] + df["Defense"] + df["Sp. Atk"] + df["Sp. Def"] + df["Speed"] 
df.head()
```

```python
# drop a column
df = df.drop(columns = ["total"])
```

### Filtering Data

```python
df.loc[df["Type 1"] == "Grass"]
```

```python
df.loc[(df["Type 1"] == "Grass") | (df["Type 2"] == "Posion")]
```

```python
df.loc[df["Type 1"].str.contains("fire|grass", flags = re.I, regex=True)]
```

```python
df.loc[df["Name"].str.contains("^pi[a-z]*", flags = re.I, regex=True)]
```

### Conditional Changes

```python
# change all "Fire" to "Flamer"
df.loc[df["Type 1"] == "Fire", "Type 1"] = "Flamer"
```

```python
# if Attach > 100 --> then the "Generation" & "Legendary" columns would be changed to "Test1", "Test2"
df.loc[df["Attack"] > 100, ["Generation", "Legendary"]] = ["Test1", "Test2"]
print(df["Generation"].unique())
print(df["Legendary"].unique())
```

### Aggregate Statistics (Groupby)

```python
df.groupby(["Type 1"]).mean()
```

```python
df.groupby(["Type 1"]).sum()
```

```python
df['count'] = 1
df.groupby(["Type 1"]).count()
print(df.groupby(["Type 1"]).count()['count'])
```

```python
df.groupby(["Type 1", "Type 2"]).count()['count']
```

```python
df.groupby(["Type 1"]).mean().sort_values("Defense", ascending = False)
```

### working with large amount of data

```python
new_df = pd.DataFrame(columns = df.columns)

for df in pd.read_csv("pokemon_data.csv", chunksize=5):
    results = df.groupby(["Type 1"]).count() # do analysis on the results for a small chunk
    new_df = pd.concat([new_df, results]) # concatenate the reuslts to a new dataframe
```
